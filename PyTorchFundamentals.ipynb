### Random Tensors

```python
# Create a random tensor of size (3,4)
random_tensor = torch.rand(3,4)
```

```python
random_tensor
```

```python
# create a random tensor with similar shape to an image tensor
random_image_tensor = torch.rand(size=(224,224,3))
random_image_tensor.shape, random_image_tensor.ndim
```

```python
# Transpose the dimensions to match the image format (H, W, C)
random_image_np = random_image_tensor.numpy()
# Display the image using matplotlib
plt.imshow(random_image_np)
plt.axis('off')  # Turn off axis labels
plt.show()
```

## Zeros and Ones

```python
# Create a tensor of all zeros
zeros = torch.zeros(size=(3,4))
```

```python
zeros
```

```python
# Create a tensor of all ones
ones = torch.ones(size=(3,4))
```

```python
ones, ones.dtype
```

## Creating a range of tensors and tensors-like

```python
# use torch.arange
one_to_ten = torch.arange(0,10)
one_to_ten
```

```python
# creating tensors like
ten_zeros = torch.zeros_like(input=one_to_ten)
ten_zeros
```

## Tensor datatypes

```python
#float32
float_32_tensor = torch.tensor([1.0, 2.0, 3.0],
                               dtype=torch.float32, # what datatype is tensor
                               device='cuda',
                               requires_grad=False # whether or not to track gradients with this tensor operation
                               )
```

```python
float_32_tensor
```

```python
float_16_tensor = float_32_tensor.type(torch.float16)
```
